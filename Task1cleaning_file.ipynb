```python
import pandas as pd
```

```python
#LOAD DATASET
df = pd.read_csv("netflix_raw.csv")
```

```python
df.head()
```

```python
#Checking basic information about the dataset and the missing values in each column
df.info()
df.isnull().sum()
```

```python
# Replace missing values in director, cast, and country with "Unknown"
df['director'] = df['director'].fillna("Unknown")
df['cast'] = df['cast'].fillna("Unknown")
df['country'] = df['country'].fillna("Unknown")
```

```python
#Dropping the rows where rating and duration is missing.
df= df.dropna(subset=['rating', 'duration'])
```

```python
#Converting the date_added column to datetime format
df['date_added']= pd.to_datetime(df['date_added'], errors='coerce')

# Drop rows where date conversion failed
df = df.dropna(subset=['date_added'])
```

```python
# Confirming all mising vales are handled
df.isnull().sum()
```

```python
# Checking number of duplicate rows
df.duplicated().sum()

# Remove duplicate rows if any
df= df.drop_duplicates()
```

```python
# Converting the 'type' column to lowercase for consistency
df['type'] = df['type'].str.lower()

# Standardize country names (capitalize properly)
df['country'] = df['country'].str.title()

# Standardize rating column to uppercase
df['rating'] = df['rating'].str.upper()
```

```python
# Make column names lowercase and replace spaces with underscores 
df.columns = df.columns.str.lower().str.replace(" ", "_")

# Check updated column names
df.columns
```

```python
# Checking final data types
df.info()
```

```python
# Final Check
print("Missing values:\n", df.isnull().sum())
print("Duplicate rows:", df.duplicated().sum())
print("Final dataset shape:", df.shape)
```

```python
# Export cleaned dataset
df.to_csv("netflix_cleaned.csv", index=False)

print("Data Cleaning Completed Successfully")
```
